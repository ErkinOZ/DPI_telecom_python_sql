### Metrics, here I show how I understand the basic metrics for telecom companies, in visual data and numbers it may be that the difference is not strong, because here are generated data

```python
import mysql.connector  # or import pymysql
import pandas as pd

#Database connection parameters
db_config = {
    "host": "192.168.1.164",  # Specify your host
    "user": "######",           # Specify your username
    "password": "####",  # Specify your password
    "database": "raw_data"      # Specify your database name
}

#Connecting to MySQL database
connection = mysql.connector.connect(**db_config)  # or pymysql.connect(**db_config)

#SQL query to retrieve data
query = "SELECT * FROM raw_data.dpi_sessions;"  # Name of your table

#Reading data into a DataFrame
df = pd.read_sql(query, connection)

#Closing the connection
connection.close()

#Displaying the first 5 rows of the DataFrame
print(df.head())
```

```python
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Convert 'start_time' to datetime if not already
df['start_time'] = pd.to_datetime(df['start_time'])

# Group by day (or any time period)
df['day'] = df['start_time'].dt.date
traffic_by_day = df.groupby('day')['traffic_volume'].sum()

# Plotting
plt.figure(figsize=(10, 6))
traffic_by_day.plot(kind='line')
plt.title('Total Traffic Volume by Day')
plt.xlabel('Date')
plt.ylabel('Traffic Volume (bytes)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_traffic_per_session = df['traffic_volume'].mean()
print(f"Average Traffic per Session: {avg_traffic_per_session} bytes")
```

```python
avg_session_duration = df['duration'].mean()
print(f"Average Session Duration: {avg_session_duration} seconds")
```

```python
traffic_by_device = df.groupby('device_type')['traffic_volume'].sum()

# Plotting
traffic_by_device.plot(kind='bar', figsize=(4, 2), color='skyblue')
plt.title('Traffic Volume by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Traffic Volume (bytes)')
plt.xticks(rotation=0)
plt.show()
```

```python
traffic_by_protocol = df.groupby('protocol')['traffic_volume'].sum()

# Plotting
traffic_by_protocol.plot(kind='pie', figsize=(2, 2), autopct='%1.1f%%', colors=['lightcoral', 'skyblue', 'gold', 'green'])
plt.title('Traffic Volume by Protocol')
plt.ylabel('')
plt.show()
```

```python
traffic_direction = df['traffic_direction'].value_counts()

# Plotting
traffic_direction.plot(kind='bar', figsize=(2, 2), color=['lightgreen', 'lightcoral'])
plt.title('Traffic Direction (Uploaded vs Downloaded)')
plt.xlabel('Traffic Direction')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
traffic_by_region = df.groupby('region')['traffic_volume'].sum()

# Plotting
traffic_by_region.plot(kind='bar', figsize=(20, 3), color='lightblue')
plt.title('Traffic Volume by Region')
plt.xlabel('Region')
plt.ylabel('Traffic Volume (bytes)')
plt.xticks(rotation=45)
plt.show()
```

```python
risk_category_distribution = df['risk_category'].value_counts()

# Plotting
risk_category_distribution.plot(kind='pie', figsize=(2, 2), autopct='%1.1f%%', colors=['yellowgreen', 'orange', 'red'])
plt.title('Risk Level Distribution')
plt.ylabel('')
plt.show()
```

```python
traffic_by_content = df.groupby('content_type')['traffic_volume'].sum()

# Plotting
traffic_by_content.plot(kind='bar', figsize=(2, 2), color='lightgreen')
plt.title('Traffic Volume by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Traffic Volume (bytes)')
plt.xticks(rotation=0)
plt.show()
```
